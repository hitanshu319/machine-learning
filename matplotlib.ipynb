```
x=[2,3]
y=[9,5]
x1=[1,2,3,4]
y1=[5,6,7,8]
plt.xlabel("time")
plt.ylabel("Speed")
plt.plot(x,y,label="water")
plt.plot(x1,y1,label="sand")
plt.grid(color='green')
plt.legend()
plt.show()
```

```
x=[2,3]
y=[9,5]
x1=[1,2,3,4]
y1=[5,6,7,8]
plt.xlabel("time")
plt.ylabel("Speed")
plt.plot(x,y,label="water")
plt.plot(x1,y1,label="sand")
plt.grid(color='green')
plt.legend()
plt.xlim(0,15)
plt.ylim(0,15)
plt.show()
```

```
x=[2,3]
y=[9,5]
x1=[1,2,3,4]
y1=[5,6,7,8]
plt.xlabel("time")
plt.ylabel("Speed")
plt.plot(x,y,label="water")
plt.plot(x1,y1,label="sand")
plt.bar(x1,y1)
plt.grid(color='green')
plt.legend()
plt.xlim(0,15)
plt.ylim(0,15)
plt.show()
```

```
import matplotlib.pyplot as plt
x=[2,3]
y=[9,5]
x1=[1,2,3,4]
y1=[5,6,7,8]
plt.xlabel("time")
plt.ylabel("Speed")
plt.plot(x,y,label="water")
plt.plot(x1,y1,label="sand")
plt.bar(x1,y1)
plt.bar(x,y)
plt.grid(color='green')
plt.legend()
plt.xlim(0,15)
plt.ylim(0,15)
plt.show()
```
